```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_cols = merged_df.select_dtypes(include=['float64', 'int64'])

# Handle null values (drop them or fill with 0/NaN if needed)
numeric_cols_cleaned = numeric_cols.dropna()  # Drop rows with null values

# Compute correlation matrix
correlation_matrix = numeric_cols_cleaned.corr()

# Visualize the correlation matrix
plt.figure(figsize=(18, 14))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar_kws={'shrink': 0.6},
    annot_kws={"size": 7}
)
plt.title("Correlation Matrix")
plt.show()

# Focus on correlation between GDP and Healthcare Expenditure
if 'GDP' in numeric_cols_cleaned.columns and 'Healthcare_Expenditure' in numeric_cols_cleaned.columns:
    gdp_healthcare_corr = numeric_cols_cleaned['GDP'].corr(numeric_cols_cleaned['Healthcare_Expenditure'])
    print("\nCorrelation between GDP and Healthcare Expenditure:", gdp_healthcare_corr)
```

```python
# Compute Average GDP and Healthcare Expenditure for each country
numeric_cols_cleaned['Avg_GDP'] = numeric_cols_cleaned[[col for col in numeric_cols_cleaned.columns if col.startswith('GDP_')]].mean(axis=1)
numeric_cols_cleaned['Avg_Health'] = numeric_cols_cleaned[[col for col in numeric_cols_cleaned.columns if col.startswith('Health_')]].mean(axis=1)

# Compute correlation between Average GDP and Average Healthcare Expenditure
avg_corr = numeric_cols_cleaned['Avg_GDP'].corr(numeric_cols_cleaned['Avg_Health'])
print("\nCorrelation between Average GDP and Average Healthcare Expenditure:", avg_corr)

# Plot the relationship between Avg_GDP and Avg_Health
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Avg_GDP', y='Avg_Health', data=numeric_cols_cleaned)
plt.title("Scatter Plot: Average GDP vs Average Healthcare Expenditure")
plt.xlabel("Average GDP")
plt.ylabel("Average Healthcare Expenditure")
plt.grid(True)
plt.show()
```

```python
# Compute year-wise correlations
correlations = {}
for year in range(2003, 2023):
    gdp_col = f'GDP_{year}'
    health_col = f'Health_{year}'
    if gdp_col in numeric_cols_cleaned.columns and health_col in numeric_cols_cleaned.columns:
        correlations[year] = numeric_cols_cleaned[gdp_col].corr(numeric_cols_cleaned[health_col])

# Print correlations for each year
print("\nYear-wise Correlations:")
for year, corr in correlations.items():
    print(f"{year}: {corr}")
```

```python
# Visualize year-wise correlations
plt.figure(figsize=(12, 6))
years = list(correlations.keys())
correlation_values = list(correlations.values())

plt.plot(years, correlation_values, marker='o', linestyle='-', color='b')
plt.title('Year-wise Correlation Between GDP and Healthcare Expenditure')
plt.xlabel('Year')
plt.ylabel('Correlation Coefficient')
plt.axhline(y=0, color='r', linestyle='--', linewidth=1)
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

# Dictionary to store correlations and p-values
correlation_pvalues = {}

# Loop through the years to calculate correlation and p-value
for year in range(2003, 2022):
    gdp_col = f'GDP_{year}'
    health_col = f'Health_{year}'
    if gdp_col in numeric_cols_cleaned.columns and health_col in numeric_cols_cleaned.columns:
        corr, p_value = pearsonr(numeric_cols_cleaned[gdp_col], numeric_cols_cleaned[health_col])
        correlation_pvalues[year] = (corr, p_value)

# Display the results
print("\nYear-wise Correlations and P-values:")
print(f"{'Year':<8}{'Correlation':<15}{'P-value'}")
for year, (corr, p_value) in correlation_pvalues.items():
    print(f"{year:<8}{corr:<15.4f}{p_value:.4e}")
```

```python
# Extract data for visualization
years = list(correlation_pvalues.keys())
correlations = [correlation_pvalues[year][0] for year in years]
p_values = [correlation_pvalues[year][1] for year in years]

# Plot correlations
fig, ax1 = plt.subplots(figsize=(8, 4))

# Correlation line plot
ax1.plot(years, correlations, color='blue', marker='o', label="Correlation")
ax1.set_xlabel('Year')
ax1.set_ylabel('Correlation', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.axhline(0, color='red', linestyle='--', linewidth=1, label='Zero Correlation')

# P-value line plot on secondary axis
ax2 = ax1.twinx()
ax2.plot(years, p_values, color='green', marker='x', linestyle='--', label="P-value")
ax2.set_ylabel('P-value', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.axhline(0.05, color='purple', linestyle='-.', linewidth=1, label='Significance Threshold (0.05)')

# Add legends and title
fig.legend(loc="upper right", bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)
plt.title('Year-wise Correlations and P-values Between GDP and Healthcare Expenditure')
plt.grid(True)
plt.show()
```

```python
# Subset data for key years
key_years = [2007, 2008, 2020]
gdp_health_key_years = numeric_cols_cleaned[
    [f'GDP_{year}' for year in key_years] + [f'Health_{year}' for year in key_years]
]

# Rename columns for clarity
gdp_health_key_years.columns = ['GDP_2007', 'GDP_2008', 'GDP_2020',
                                'Health_2007', 'Health_2008', 'Health_2020']

# Create boxplots for GDP and Healthcare
plt.figure(figsize=(10, 6))
gdp_health_key_years[['GDP_2007', 'GDP_2008', 'GDP_2020']].boxplot()
plt.title("GDP Distribution for Key Years (2007, 2008, 2020)")
plt.ylabel("GDP")
plt.show()

plt.figure(figsize=(10, 6))
gdp_health_key_years[['Health_2007', 'Health_2008', 'Health_2020']].boxplot()
plt.title("Healthcare Expenditure Distribution for Key Years (2007, 2008, 2020)")
plt.ylabel("Healthcare Expenditure")
plt.show()
```

```python
# Identify outliers for GDP and Healthcare
outliers = {}
for year in key_years:
    gdp_col = f'GDP_{year}'
    health_col = f'Health_{year}'
    outliers[year] = {
        "GDP_Outliers": numeric_cols_cleaned[
            numeric_cols_cleaned[gdp_col] > gdp_health_key_years[gdp_col].quantile(0.75) +
            1.5 * (gdp_health_key_years[gdp_col].quantile(0.75) - gdp_health_key_years[gdp_col].quantile(0.25))
        ]['Country_Name'],
        "Health_Outliers": numeric_cols_cleaned[
            numeric_cols_cleaned[health_col] > gdp_health_key_years[health_col].quantile(0.75) +
            1.5 * (gdp_health_key_years[health_col].quantile(0.75) - gdp_health_key_years[health_col].quantile(0.25))
        ]['Country_Name']
    }

# Print outliers for each year
for year, outlier_data in outliers.items():
    print(f"Outliers for {year}:")
    print("GDP Outliers:")
    print(outlier_data["GDP_Outliers"].values)
    print("Healthcare Outliers:")
    print(outlier_data["Health_Outliers"].values)
    print("-" * 50)
```

```python
# Define the Region Mapping
region_mapping = {
    "United States": "North America",
    "Canada": "North America",
    "Mexico": "North America",
    "India": "Asia",
    "Japan": "Asia",
    "China": "Asia",
   }

# Make a copy of the DataFrame to avoid modifying the original
numeric_cols_cleaned_copy = numeric_cols_cleaned.copy()

# Map Regions to Countries
numeric_cols_cleaned_copy['Region'] = numeric_cols_cleaned_copy['Country_Name'].map(region_mapping)

# Convert GDP and Healthcare columns to numeric
gdp_cols = [col for col in numeric_cols_cleaned_copy.columns if col.startswith('GDP_')]
health_cols = [col for col in numeric_cols_cleaned_copy.columns if col.startswith('Health_')]

for col in gdp_cols + health_cols:
    numeric_cols_cleaned_copy[col] = pd.to_numeric(numeric_cols_cleaned_copy[col], errors='coerce')

# Drop rows with invalid Region or missing numeric values
numeric_cols_cleaned_copy = numeric_cols_cleaned_copy.dropna(subset=['Region'] + gdp_cols + health_cols)

# Compute Regional Averages
regional_data = numeric_cols_cleaned_copy.groupby('Region').mean(numeric_only=True).reset_index()

# Add Average GDP and Healthcare Expenditure Columns
regional_data['Avg_GDP'] = regional_data[gdp_cols].mean(axis=1)
regional_data['Avg_Health'] = regional_data[health_cols].mean(axis=1)

regional_gdp_health = regional_data[['Region', 'Avg_GDP', 'Avg_Health']]

# Display the Resulting Regional Data
print("\nRegional Averages for GDP and Healthcare Expenditure:")
print(regional_gdp_health)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the average GDP and Healthcare expenditure per region
plt.figure(figsize=(6, 4))
sns.barplot(x='Region', y='Avg_GDP', data=regional_data, label='Average GDP', color='cyan')
sns.barplot(x='Region', y='Avg_Health', data=regional_data, label='Average Healthcare Expenditure', color='pink')
plt.title('Average GDP and Healthcare Expenditure by Region')
plt.xlabel('Region')
plt.ylabel('Expenditure (in respective units)')
plt.legend(['Average GDP', 'Average Healthcare Expenditure'])
plt.grid(axis='y')
plt.show()
```
